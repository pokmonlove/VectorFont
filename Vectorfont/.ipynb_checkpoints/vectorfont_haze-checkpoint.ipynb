## 2. 폰트 로드

```python
f = font.open("fonts/NanumSquare_0.ttf")
print("폰트명:", f.name)
print("Density:", f.density)
```

## 3. 자주 사용할 함수 정의

```python
def glyphcommands(f, ch):
    return Path([copy(cmd) for cmd in f.glyph(f.charmap[ord(ch)])])

def advancefor(f, ch):
    return f.advances[f.charmap[ord(ch)]]
```

```python
from bezmerizing import PolylineList
```

```python
from itertools import chain
#문자열의 각 문자에 대한 경로를 결합하는 경로를 반환하는 함수
def combine_path(f, s):
    text_paths = []
    cx = 0
    for ch in s:
        glyph_path = glyphcommands(f, ch).translate(cx, 0)
        text_paths.append(glyph_path)
        cx += advancefor(f, ch)
    combined = Path(list(chain(*text_paths)))
    return combined
```

## 4. path를 polyline으로 바꾸기

```python
qf_path = combine_path(f, "아지랑이")
qf_plist = qf_path.to_polyline_list(samples_per=24, resample_polylines=True)
```

```python
#len(qf_path.commands)
```

```python
page = document(300, 100, 'pt').addpage()
pen = shape().stroke(rgb(40, 40, 40)).fill(rgb(240, 240, 240))
factor = 74 / f.density
page.place(pen.path(qf_path.scale(factor).translate(20, 74)))
show(page)
```

▼ 각 폴리라인마다 PolylineList 반복하는 방법 (▲방법은 .map() 메서드를 사용해 목록의 각 Polyline에 동일한 기능을 적용함)

mod = []
for pline in qf_plist:
    verts = []
    for item in pline.vertices:
        verts.append([item[0] + np.sin(item[1] * 0.01)*20,
                      item[1] + np.cos(item[0] * 0.01) * 20])
    new_pline = Polyline(verts)
    mod.append(new_pline)
mod_plist = PolylineList(mod)

page = document(400, 90, 'pt').addpage()
pen = shape().stroke(rgb(40, 40, 40)).fill(rgb(240, 240, 240))
factor = 72 / f.density
cmds = []
page.place(pen.path(mod_plist.to_path().scale(factor).translate(20, 74)))
show(page)
